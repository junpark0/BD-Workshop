```python
### Import your libraries first!
import numpy as np
import pandas as pd
import seaborn as sns
```

### Basic Pandas Syntax

```python
### Pull the data from the csv file and assign it to 'nba' variable
### .head() function returns the top 5 entries of the dataframe
### .head(x) will return the top 'x' entries of the dataframe

nba = pd.read_csv('nba18-19.csv')
nba.head()
```

```python
nba
```

```python
### Call a specific column by using the format: dataframe['COLUMN NAME']  -->  Returns a SERIES

nba['Player']
```

```python
### Call a specific column by using the format: dataframe[['COLUMN NAME']]  -->  Returns a DATAFRAME

nba[['Player']]
```

```python
### Call multiple columns by using the format: dataframe[['FIRST COLUMN', 'SECOND COLUMN', 'THIRD COLUMN', etc]]

nba[['Player', 'Pos']]
```

### String Methods

```python
### Finding the entries with texts starting with 'A'
### Use the format:
### dataframe['COLUMN'].str.startswith('A')
### Similar format if we want to find entries with texts ending with 'A'
### Use the format:
### dataframe['COLUMN'].str.endswith('A')
### Returns a SERIES of True or False

nba['Player'].str.startswith('Y')
```

```python
### Lowercase all of the entries in a dataframe column
### Use the format:
### dataframe['COLUMN'].str.lower()
### Similar format to uppercase all of the entries in a dataframe column
### Use the format:
### dataframe['COLUMN'].str.upper()

nba['Player'].str.lower()
```

```python
### Split an entry for the entire column using the format: dataframe['COLUMN'].str.split('x')
### 'x' is whatever you want to split the text on
### if a entry is 'Apple_Pie' and you want to split into 'Apple' and 'Pie', use:
### dataframe['COLUMN'].str.split('_')

nba['Player'].str.split('\\')
```

```python
nba['Player'].str.split('\\')[0]
```

```python
nba['Player'].str.split('\\').str[0]
```

### DataFrame Manipulation

```python
### Add a new column to the existing dataframe by using the format: 
### dataframe['NEW COLUMN NAME'] = whatever information you want to put there

nba['Player_Name'] = nba['Player'].str.split('\\').str[0]
nba
```

```python
### Perform calculations columns
### mean(), sum(), median(), etc.

nba['Age'].mean()
```

```python
### Use dataframe.unique() function to find all the unique values in a specific column

nba['Tm'].unique()
```

```python
### Filter dataframes by a specific key word you want
### Use the format: 
### dataframe[dataframe['COLUMN'] == 'SPECIFIC WORD']
### "Return the dataframe where dataframe['COLUMN'] is equal to the specific word"

nba[nba['Pos'] == 'C']
```

```python
### Filter dataframes by a specific key wordS you want
### Use the format: 
### dataframe[dataframe['COLUMN'].isin(['WORD !', 'WORD 2', 'WORD 3'])]
### "Return the dataframe where dataframe['COLUMN'] contains any WORD 1, WORD 2, or WORD 3"

nba[nba['Tm'].isin(['GSW', 'MIA'])].head()
```

```python
### Filter dataframes by a specific values you want
### Use the format: 
### dataframe[dataframe['COLUMN'] >= VALUE]
### "Return the dataframe where dataframe['COLUMN'] is greater than or equal to a certain value"

nba[nba['Age'] < 20]
```

```python
### We can store this dataframe into a new variable and call it whenever

under_20 = nba[nba['Age'] < 20]
under_20
```

```python
### Find the different unique NBA positions listed in the DataFrame using .unique()

nba['Pos'].unique()
```

```python
### Filter dataframes by a specific string you want
### Use the format: 
### dataframe[dataframe['COLUMN'].str.contains('STRING OF CHARACTERS')]
### "Return the dataframe where dataframe['COLUMN'] entry contains 'STRING OF CHARACTERS' in its entries

nba[nba['Pos'].str.contains('F')]
```

```python
### Earlier we did nba['Player'].str.startswith('Y')] and returned a series of True and False for the entire
### 'Player' column to see which entries started with 'Y'
### What if we wanted to actually see the players with names that start with 'Y'?

nba[nba['Player'].str.startswith('Y')]
```

### Groupby Method

<img src="groupby.png"/>

```python
### Group the information from the dataframe by specific column
### THIS DOES NOT RETURN ANYTHING

nba.groupby('Tm')
```

```python
### But if you want to see what happens when you groupby a column,
### copy and paste the following code
### In this case, we are grouping by the different NBA teams, so we can see the statistics per team in each sub-table

for n, g in nba.groupby('Tm'):
    print(f"Name: {n}") 
    display(g)
```

```python
### What other columns are good for grouping by?

for n, g in nba.groupby('Age'):
    print(f"Name: {n}") 
    display(g)
```

```python
### After the dataframe.groupby('COLUMN'), we need to add an aggregate function if we want our code to return a dataframe
### Aggregate functions can be simple or complicated, but we will only focus on simple aggregate functions for now
### In the following code, we are applying the mean() function to every column (if applicable) by each team
### mean() function gets applied to every column in each sub-table,
### and returns one row per sub-table that will be added to the final table

nba.groupby('Tm').mean()
```

```python
nba.groupby('Tm').sum()
```

```python
### If we want to see how many times each entry comes out, use the format:
### dataframe.groupby('COLUMN').size()  -->  returns a SERIES

nba.groupby('Tm').size()
```

```python
### If we want to return a dataframe, add: .reset_index(name = 'counts')

nba.groupby('Tm').size().reset_index(name = 'counts')
```

```python
### We can also numerically sort our dataframe using:
### .sort_values(by = 'COLUMN YOU WILL BE BASING YOUR SORT FUNCTION ON')
### This will return the dataframe in increasing order from top to bottom

nba.groupby('Tm').size().reset_index(name = 'counts').sort_values(by = 'counts')
```

```python
### If we want to sort our values by decreasing order (bigger values on top and smaller values on bottom), use:
### .sort_values(by = 'COLUMN YOU WILL BE BASING YOUR SORT FUNCTION ON', ascending=False)

nba.groupby('Tm').size().reset_index(name = 'counts').sort_values(by = 'counts', ascending=False)
```

```python
### We can also group our dataframe with two columns:
### Use the format: dataframe.groupby(['COLUMN 1', 'COLUMN 2'].aggfunc()

nba.groupby(['Tm', 'Pos']).mean()
```

```python
### We can select a specific column from our groupby dataframe with the following format:
### In this case, we are grouping by team and position, then using the mean function on the available columns,
### then selecting the 'PTS' column to return (in a SERIES format)

nba.groupby(['Tm', 'Pos']).mean()['PTS']
```

```python
### Return a dataframe using [['COLUMN']]  --> double brackets

nba.groupby(['Tm', 'Pos']).mean()[['PTS']]
```

### Visualization

```python
### Histogram/Density Curve plot for PTS 
### Histogram/Density Curves must be for a quantitative (numerical) variable
### sns.distplot(dataframe['COLUMN']);

sns.distplot(nba['PTS']);
```

```python
### If you want the lines for the underlying histogram
### add this to the code
### sns.distplot(dataframe['COLUMN'], hist_kws=dict(ec='W'));

sns.distplot(nba['PTS'], hist_kws=dict(ec='W'));
```

```python
### Countplot for number of players per NBA team
### Countplot is useful for plotting the counts of a qualitative (categorical) variable
### sns.countplot(dataframe['COLUMN'])

sns.countplot(nba['Tm']);
```

```python
### Rotate x-axis labels using the following code:
### messed_up_x_axis_graph = sns.countplot(dataframe['COLUMN']);
### messed_up_x_axis_graph;
### messed_up_x_axis_graph.set_xticklabels(messed_up_x_axis_graph.get_xticklabels(), rotation = 90);

team_graph = sns.countplot(nba['Tm'])
team_graph; 
team_graph.set_xticklabels(team_graph.get_xticklabels(), rotation = 90);
```

```python
### Smaller dataset for graphing purposes

shortened_nba = nba.sample(10)
shortened_nba
```

```python
### Barplot for players and their respective points
### sns.barplot(dataframe['COLUMN 1'], dataframe['COLUMN 2'])
### COLUMN 1 should be qualitative-categorical (x-axis)
### COLUMN 2 should be quantitative-numerical (y-axis)

player_pts_graph = sns.barplot(shortened_nba['Player_Name'], shortened_nba['PTS']);
player_pts_graph;
player_pts_graph.set_xticklabels(player_pts_graph.get_xticklabels(), rotation = 90);
```

```python
### Box plots are useful for quantitative (numerical) distribution
### Lower Quartile, Median, Upper Quartile, Whiskers, Outliers
### sns.boxplot(dataframe['COLUMN']);

sns.boxplot(nba['PTS']);
```

```python
### Violin plots are useful for quantitative (numerical) distribution
### Similar to Box plot, but also shows smoothed density curves
### "Width of the box" now has meaning
### sns.violinplot(dataframe['COLUMN']);

sns.violinplot(nba['PTS']);
```

```python
### Filtering the data so we only work with NBA teams in California
### LAL, LAC, SAC, GSW

ca_teams = nba[nba['Tm'].isin(['LAL', 'LAC', 'SAC', 'GSW'])]
ca_teams
```

```python
### Boxplots and Violinplots are useful for comparing multiple distributions
### Use the format:
### sns.___plot(data = dataframe, x = 'COLUMN 1', y = 'COLUMN 2');
### COLUMN 1 should be qualitative (categorical)
### COLUMN 2 should be quantitative (numerical)

sns.violinplot(data = ca_teams, x = 'Tm', y = 'PTS');
```

```python
sns.boxplot(data = ca_teams, x = 'Tm', y = 'PTS')
```
